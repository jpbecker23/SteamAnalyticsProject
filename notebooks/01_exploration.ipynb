# 🎮 Steam Games Analysis - Exploração de Dados

Este notebook documenta a análise exploratória inicial (EDA) do dataset da Steam. O objetivo é entender a estrutura dos dados, realizar a limpeza necessária e validar as hipóteses para o dashboard final.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações básicas
pd.set_option('display.max_columns', None)
sns.set_style("darkgrid")
```

## 1. Carregamento e Correção
Durante a implementação, identificamos que o header do CSV original estava desalinhado (39 colunas no header vs 40 nos dados). Vamos carregar usando a estrutura corrigida.

```python
CORRECT_COLUMNS = [
    'AppID', 'Name', 'Release date', 'Estimated owners', 'Peak CCU',
    'Required age', 'Price', 'Discount', 'DLC count',
    'About the game', 'Supported languages', 'Full audio languages',
    'Reviews', 'Header image', 'Website', 'Support url', 'Support email',
    'Windows', 'Mac', 'Linux', 'Metacritic score', 'Metacritic url',
    'User score', 'Positive', 'Negative', 'Score rank', 'Achievements',
    'Recommendations', 'Notes', 'Average playtime forever',
    'Average playtime two weeks', 'Median playtime forever',
    'Median playtime two weeks', 'Developers', 'Publishers',
    'Categories', 'Genres', 'Tags', 'Screenshots', 'Movies'
]

df = pd.read_csv('../data/raw/games.csv', header=0, names=CORRECT_COLUMNS)
print(f"Dataset carregado com {df.shape[0]} linhas e {df.shape[1]} colunas.")
```

## 2. Limpeza de Dados
Vamos focar nas colunas essenciais: `Name`, `Price`, `Positive`, `Negative`.

```python
# Conversão de Colunas Numéricas
df['Positive'] = pd.to_numeric(df['Positive'], errors='coerce').fillna(0)
df['Negative'] = pd.to_numeric(df['Negative'], errors='coerce').fillna(0)
df['Total_Reviews'] = df['Positive'] + df['Negative']

# Cálculo do Score (%)
df['Review_Score'] = (df['Positive'] / df['Total_Reviews'] * 100).fillna(0).round(1)

# Extração de Ano
df['Release_Date_DT'] = pd.to_datetime(df['Release date'], errors='coerce')
df['Release_Year'] = df['Release_Date_DT'].dt.year.fillna(0).astype(int)

df[['Name', 'Total_Reviews', 'Review_Score', 'Release_Year']].head()
```

## 3. Exploração Inicial
Quais são os jogos mais populares de todos os tempos?

```python
top_20 = df.nlargest(20, 'Total_Reviews')

plt.figure(figsize=(10, 8))
sns.barplot(data=top_20, x='Total_Reviews', y='Name', palette='viridis')
plt.title('Top 20 Jogos por Número de Avaliações')
plt.show()
```

## 4. Conclusão da EDA
Os dados estão prontos para serem usados no dashboard Streamlit. O fix de colunas foi essencial para garantir a integridade dos nomes dos jogos.
